```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imshow
from skimage.color import rgb2gray, rgb2hsv
from skimage.morphology import area_opening
from skimage.exposure import histogram
from skimage.filters import threshold_otsu
from PIL import Image
```

```python
import os
path = 'Dataset'
```

```python
#Proses Thresholding
def masked_image(image, mask):
    r = image[:,:,0] * mask
    g = image[:,:,1] * mask
    b = image[:,:,2] * mask
    return np.dstack([r,g,b])

for filename in os.listdir(path):
    for imagename in os.listdir(os.path.join(path, filename)):
        image = imread(os.path.join(path, filename, imagename))
        gray = rgb2gray(image)
        thresh = threshold_otsu(gray)
        masking  = gray < thresh
        filtered = masked_image(image, masking)
        image = Image.fromarray(filtered)
        image = image.resize((260, 146))
        image = np.asarray(image)
        
        rows, cols, _ = filtered.shape
        
        data_piksel = []
        for i in range(rows):
            for j in range(cols):
                k = filtered[i,j]
                r,g,b = k
                sum = r + g + b
                if sum > 0:
                    norm_r = int(r) / (int(r) + int(g) + int(b))
                    norm_g = int(g) / (int(r) + int(g) + int(b))
                    norm_b = int(b) / (int(r) + int(g) + int(b))
                    nilai_v = ''
                    #Normalisasi V
                    if norm_r > norm_g and norm_r > norm_b:
                        maks = float(norm_r)
                        nilai_v = 'r'
                    elif norm_g > norm_r and norm_g > norm_b:
                        maks = float(norm_g)
                        nilai_v = 'g'
                    else:
                        maks = float(norm_b)
                        nilai_v = 'b'
                    #print(maks)
                    #Normalisasi S
                    if norm_r < norm_g and norm_r < norm_b:
                        minim = norm_r
                    elif norm_g < norm_r and norm_g < norm_b:
                        minim = norm_g
                    else:
                        minim = norm_b
                    #Menghitung S
                    if maks == 0:
                        nilai_s = float(0)
                    else: #Nilai V > 0
                        nilai_s = maks - (minim/maks)
                    #print(nilai_s)
                    #Normalisasi H
                    if nilai_s == 0:
                        nilai_h = 0
                    elif nilai_v == 'r':
                        nilai_h = 60 * (norm_g - norm_b) / nilai_s * maks
                    elif nilai_v == 'g':
                        nilai_h = 60 * (2 + (norm_b - norm_r) / nilai_s * nilai_v)
                    elif nilai_v == 'b':
                        nilai_h = 60 * (4 + (norm_r - norm_g) / nilai_s * nilai_v)
                    print(nilai_h, nilai_s, nilai_v)
        break
        
        #print(data_piksel)
        
        plt.title(filename)
        plt.imshow(image)
        plt.show()
        break
```
